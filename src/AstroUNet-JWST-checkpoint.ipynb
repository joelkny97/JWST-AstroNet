```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class AstroUNet(nn.Module):
    def __init__(self, exp_injection=True):
        super().__init__()
        self.exp_injection = exp_injection

        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, padding=1),
                nn.LeakyReLU(0.2),
                nn.Conv2d(out_c, out_c, 3, padding=1),
                nn.LeakyReLU(0.2)
            )

        self.enc1 = conv_block(1, 32)
        self.enc2 = conv_block(32, 64)
        self.enc3 = conv_block(64, 128)
        self.enc4 = conv_block(128, 256)
        self.enc5 = conv_block(256, 512)

        self.pool = nn.MaxPool2d(2)

        self.dec6 = conv_block(512 + 256 + (1 if exp_injection else 0), 256)
        self.dec7 = conv_block(256 + 128, 128)
        self.dec8 = conv_block(128 + 64, 64)
        self.dec9 = conv_block(64 + 32, 32)

        self.final = nn.Conv2d(32, 1, kernel_size=1)

    def center_crop(self, enc_feat, target):
        _, _, h, w = enc_feat.size()
        _, _, th, tw = target.size()
        dh, dw = h - th, w - tw
        cropped = enc_feat[:, :, dh//2:h - (dh - dh//2), dw//2:w - (dw - dw//2)]
        return cropped

    def forward(self, x, exp_time=None):
        e = None
        if self.exp_injection and exp_time is not None:
            e = torch.full_like(x[:, :1], exp_time)

        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))
        e5 = self.enc5(self.pool(e4))

        # Decoder with interpolation + safe crop
        d6 = F.interpolate(e5, scale_factor=2, mode='bilinear', align_corners=False)
        e4_c = self.center_crop(e4, d6)
        if self.exp_injection:
            e_crop = self.center_crop(e, d6)
            d6 = torch.cat([d6, e4_c, e_crop], dim=1)
        else:
            d6 = torch.cat([d6, e4_c], dim=1)
        d6 = self.dec6(d6)

        d7 = F.interpolate(d6, scale_factor=2, mode='bilinear', align_corners=False)
        e3_c = self.center_crop(e3, d7)
        d7 = torch.cat([d7, e3_c], dim=1)
        d7 = self.dec7(d7)

        d8 = F.interpolate(d7, scale_factor=2, mode='bilinear', align_corners=False)
        e2_c = self.center_crop(e2, d8)
        d8 = torch.cat([d8, e2_c], dim=1)
        d8 = self.dec8(d8)

        d9 = F.interpolate(d8, scale_factor=2, mode='bilinear', align_corners=False)
        e1_c = self.center_crop(e1, d9)
        d9 = torch.cat([d9, e1_c], dim=1)
        d9 = self.dec9(d9)

        return self.final(d9)
```

```python
from astropy.io import fits
import numpy as np

def load_jwst_i2d_image(fits_path):
    with fits.open(fits_path) as hdul:
        image = hdul['SCI'].data  # 2D science image
        header = hdul['SCI'].header
        image = np.nan_to_num(image)
        
        # Normalize (1%‚Äì99% clipping)
        vmin, vmax = np.percentile(image, [1, 99])
        norm_image = np.clip((image - vmin) / (vmax - vmin), 0, 1).astype(np.float32)
        
        # Exposure time
        exp_time = header.get("EXPTIME", 1000.0)
        
    return norm_image, exp_time
```

```python
def extract_patches(image, patch_size=256, stride=256, max_patches=100):
    patches = []
    H, W = image.shape
    for i in range(0, H - patch_size + 1, stride):
        for j in range(0, W - patch_size + 1, stride):
            patch = image[i:i+patch_size, j:j+patch_size]
            if np.mean(patch) > 0.01:  # Skip blank
                patches.append(patch)
            if len(patches) >= max_patches:
                return patches
    return patches
```

```python
def simulate_noise(image, exp_time, ratio=2, ron=3, dark_current=7):
    """
    image: normalized [0, 1], assume units = e-/sec
    exp_time: in seconds
    """
    img_e = image * exp_time  # Convert to electrons

    signal = np.random.poisson(img_e / ratio)
    read_noise = np.random.normal(0, ron, image.shape)
    dark_noise = np.random.normal(0, np.sqrt(dark_current * exp_time / (3600 * ratio)), image.shape)

    noisy_e = signal + read_noise + dark_noise
    noisy_image = noisy_e / (exp_time / ratio)

    noisy_image = np.clip(noisy_image, 0, None)
    return noisy_image.astype(np.float32)
```

```python
import torch
import os

def generate_training_pairs(image, exp_time, save_dir, prefix, patch_size=256, stride=256):
    os.makedirs(save_dir, exist_ok=True)
    clean_patches = extract_patches(image, patch_size, stride)
    saved_paths = []

    for idx, clean in enumerate(clean_patches):
        noisy = simulate_noise(clean, exp_time)
        clean_tensor = torch.from_numpy(clean).unsqueeze(0)  # [1, H, W]
        noisy_tensor = torch.from_numpy(noisy).unsqueeze(0)
        
        noisy_path = os.path.join(save_dir, f"{prefix}_noisy_{idx:03d}.pt")
        clean_path = os.path.join(save_dir, f"{prefix}_clean_{idx:03d}.pt")

        torch.save(noisy_tensor, noisy_path)
        torch.save(clean_tensor, clean_path)

        saved_paths.append((noisy_path, clean_path))
    
    return saved_paths

```

```python
import torch
from torch.utils.data import Dataset
import os

class JWSTAstroDataset(Dataset):
    def __init__(self, txt_file, root_dir="./astronet_training/", transform=None):
        """
        Args:
            txt_file (str): Path to .txt file with image pair names (noisy and clean).
            root_dir (str): Directory with .pt files.
            transform (callable, optional): Optional transform to be applied.
        """
        with open(txt_file, 'r') as f:
            self.file_list = [line.strip() for line in f.readlines()]
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        noisy_file, clean_file = self.file_list[idx].split()
        noisy = torch.load(os.path.join( noisy_file))
        clean = torch.load(os.path.join( clean_file))
        if self.transform:
            noisy = self.transform(noisy)
            clean = self.transform(clean)
        return noisy, clean
```

```python
from astroquery.mast import Observations
import os

# Broader JWST filters to try
# === Filters and settings ===
jwst_filters = ["F115W", "F470N", "F480M", "F150W2", "F444W", "F162M", "F360M", "F405N", "F212N", "F182M"]
max_files_per_filter = 5
base_dir = "data"
patch_output_base = "astronet_training"


for filt in jwst_filters:
    print(f"\nüîç Searching JWST I2D data for filter: {filt}")
    output_dir = os.path.join(base_dir, filt)
    os.makedirs(output_dir, exist_ok=True)

    obs_table = Observations.query_criteria(
        obs_collection="JWST",
        instrument_name="NIRCAM/IMAGE",
        dataproduct_type="image",
        filters=[filt],
        proposal_id=4147,
        calib_level=3
    )

    if len(obs_table) == 0:
        print(f"‚ö†Ô∏è No observations found for filter {filt}")
        continue

    products = Observations.get_product_list(obs_table)
    i2d_products = Observations.filter_products(
        products,
        productSubGroupDescription="I2D",
        extension="fits"
    )

    if len(i2d_products) == 0:
        print(f"‚ö†Ô∏è No I2D products found for filter {filt}")
        continue

    i2d_products = i2d_products[:max_files_per_filter]
    manifest = Observations.download_products(i2d_products, download_dir=output_dir)
    print(f"üì• Downloaded {len(manifest)} files to {output_dir}")

    # === Now process each FITS file for training patch generation ===
    for file_path in manifest["Local Path"]:
        try:

            image, exp_time = load_jwst_i2d_image(file_path)

            prefix = os.path.splitext(os.path.basename(file_path))[0]
            save_dir = os.path.join(patch_output_base, filt)
            generate_training_pairs(image, exp_time, save_dir, prefix)
            print(f"‚úÖ Generated training pairs from {prefix}")

        except Exception as e:
            print(f"‚ùå Error processing {file_path}: {e}")
```

```python
import os
import random

def create_dataset_splits(save_dir="./astronet_training/", train_ratio=0.8, val_ratio=0.1):
    paired_files = []

    # Walk through all subdirectories and collect clean/noisy pairs
    for root, _, files in os.walk(save_dir):
        clean_files = [f for f in files if "clean" in f and f.endswith(".pt")]
        for clean_file in clean_files:
            noisy_file = clean_file.replace("clean", "noisy")
            noisy_path = os.path.join(root, noisy_file)
            clean_path = os.path.join(root, clean_file)

            if os.path.exists(noisy_path):
                paired_files.append((noisy_path, clean_path))

    print(f"üîç Found {len(paired_files)} total clean/noisy pairs")

    # Shuffle and split
    random.shuffle(paired_files)
    total = len(paired_files)
    train_end = int(train_ratio * total)
    val_end = int((train_ratio + val_ratio) * total)

    splits = {
        "train.txt": paired_files[:train_end],
        "val.txt": paired_files[train_end:val_end],
        "test.txt": paired_files[val_end:]
    }

    # Write split files with full paths (or relative to save_dir if needed)
    for split_name, split_data in splits.items():
        split_path = os.path.join(save_dir, split_name)
        with open(split_path, "w") as f:
            for noisy, clean in split_data:
                f.write(f"{noisy} {clean}\n")
        print(f"‚úÖ Saved {split_name} with {len(split_data)} pairs")

# Run it
create_dataset_splits()
```

```python
import os
import torch
import numpy as np
import pandas as pd

def compute_dataset_stats(txt_file, root_dir=""):
    # Read the list of noisy/clean filenames
    with open(txt_file, 'r') as f:
        pairs = [line.strip().split() for line in f if line.strip()]
    stats = []
    for noisy_name, clean_name in pairs:
        noisy_path = os.path.join(root_dir, noisy_name)
        clean_path = os.path.join(root_dir, clean_name)
        noisy = torch.load(noisy_path).float().squeeze().numpy()
        clean = torch.load(clean_path).float().squeeze().numpy()
        stats.append({
            "mean_noisy": noisy.mean(),
            "std_noisy": noisy.std(),
            "zero_frac_noisy": np.mean(noisy == 0),
            "mean_clean": clean.mean(),
            "std_clean": clean.std(),
            "zero_frac_clean": np.mean(clean == 0)
        })
    df = pd.DataFrame(stats)
    summary = {
        "num_samples": len(df),
        "mean_noisy_mean": df.mean_noisy.mean(),
        "mean_noisy_std":  df.mean_noisy.std(),
        "std_noisy_mean":  df.std_noisy.mean(),
        "std_noisy_std":   df.std_noisy.std(),
        "zero_noisy_avg%": df.zero_frac_noisy.mean()*100,
        "mean_clean_mean": df.mean_clean.mean(),
        "std_clean_mean":  df.std_clean.mean(),
        "zero_clean_avg%": df.zero_frac_clean.mean()*100,
    }
    return df, summary

# Compute for each split
for split in ["train", "val", "test"]:
    txt = f"./astronet_training/{split}.txt"
    df, summary = compute_dataset_stats(txt)
    print(f"\nüî∏ {split.upper()} SET")
    for k, v in summary.items():
        if isinstance(v, float):
            print(f"   {k:20s}: {v:>8.3f}")
        else:
            print(f"   {k:20s}: {v}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import os
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim
# from astrounet_dataset import JWSTAstroDataset  # from previous step
# from astrounet_model import AstroUNet          # your converted U-Net model
import numpy as np
from datetime import datetime

# === Hyperparameters ===
BATCH_SIZE = 4
NUM_EPOCHS = 100
LEARNING_RATE = 1e-4
SAVE_DIR = "./checkpoints_astrounet/"
os.makedirs(SAVE_DIR, exist_ok=True)
EXP_VAL=2.0

start_time = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
# === Data ===
# train_dataset = JWSTAstroDataset("astronet_training/train.txt")
# val_dataset = JWSTAstroDataset("astronet_training/val.txt")

# train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
# val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False)


# Dataloaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# === Model, Loss, Optimizer ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AstroUNet(exp_injection=True).to(device)
criterion = nn.L1Loss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

best_val_psnr = 0.0

# === Training Loop ===
for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0.0
    for noisy, clean in train_loader:
        noisy, clean = noisy.to(device), clean.to(device)
        exp_val = torch.tensor(EXP_VAL).to(device)  # inject exposure ratio (can vary)
        output = model(noisy, exp_val)

        loss = criterion(output, clean)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{NUM_EPOCHS}, Train Loss: {avg_loss:.4f}")

    # === Evaluation ===
    model.eval()
    psnr_vals = []
    ssim_vals = []
    with torch.no_grad():
        for noisy, clean in val_loader:
            noisy, clean = noisy.to(device), clean.to(device)
            output = model(noisy, exp_val)
            output_np = output.cpu().squeeze().numpy()
            clean_np = clean.cpu().squeeze().numpy()

            psnr_vals.append(psnr(clean_np, output_np, data_range=1.0))
            ssim_vals.append(ssim(clean_np, output_np, data_range=1.0))

    avg_psnr = np.mean(psnr_vals)
    avg_ssim = np.mean(ssim_vals)
    print(f"   üîç Val PSNR: {avg_psnr:.2f}, SSIM: {avg_ssim:.3f}")

    # === Save best model ===
    if avg_psnr > best_val_psnr:
        best_val_psnr = avg_psnr
        torch.save(model.state_dict(), os.path.join(SAVE_DIR, f"best_model_{EXP_VAL}_{start_time}.pth"))
        print("   ‚úÖ Saved new best model!")
```

```python
# Reconstruction of whole image
def extract_image_patches(image, patch_size=256, stride=128):
    H, W = image.shape
    patches = []
    positions = []
    for i in range(0, H - patch_size + 1, stride):
        for j in range(0, W - patch_size + 1, stride):
            patch = image[i:i+patch_size, j:j+patch_size]
            patches.append(patch)
            positions.append((i, j))
    return patches, positions

def reconstruct_image(patches, positions, image_shape, patch_size=256):
    reconstructed = np.zeros(image_shape)
    weight = np.zeros(image_shape)

    for patch, (i, j) in zip(patches, positions):
        reconstructed[i:i+patch_size, j:j+patch_size] += patch
        weight[i:i+patch_size, j:j+patch_size] += 1

    return reconstructed / np.maximum(weight, 1e-8)
```

```python
import torch

def denoise_full_image(image, model, exp_ratio=2.0, patch_size=256, stride=128):
    model.eval()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    norm_image = np.clip(image / np.percentile(image, 99), 0, 1).astype(np.float32)
    patches, positions = extract_image_patches(norm_image, patch_size, stride)
    denoised_patches = []

    for patch in patches:
        tensor = torch.from_numpy(patch).unsqueeze(0).unsqueeze(0).to(device)
        with torch.no_grad():
            out = model(tensor, exp_time=torch.tensor(exp_ratio).to(device))
        denoised_patches.append(out.squeeze().cpu().numpy())

    denoised_full = reconstruct_image(denoised_patches, positions, image.shape, patch_size)
    return denoised_full
```

```python
from torch.utils.data import DataLoader
import torch

def validate_dataset(dataset, zero_threshold=0.95, show_examples=False, max_examples=3):
    total = len(dataset)
    zero_count = 0
    zero_indices = []

    for idx in range(total):
        noisy, clean = dataset[idx]

        # Count nonzero pixels in clean
        nonzero = torch.count_nonzero(clean)
        total_pixels = clean.numel()
        ratio = nonzero.float() / total_pixels

        if ratio < (1 - zero_threshold):
            zero_count += 1
            zero_indices.append(idx)

            if show_examples and len(zero_indices) <= max_examples:
                import matplotlib.pyplot as plt
                plt.figure(figsize=(10, 3))
                plt.subplot(1, 2, 1)
                plt.imshow(noisy.squeeze(), cmap='gray')
                plt.title(f"Noisy Sample {idx}")
                plt.subplot(1, 2, 2)
                plt.imshow(clean.squeeze(), cmap='gray')
                plt.title(f"Clean Sample {idx} (nonzero ratio: {ratio:.2f})")
                plt.show()

    print(f"‚úÖ Dataset has {total} samples.")
    print(f"‚ö†Ô∏è {zero_count} samples ({100*zero_count/total:.1f}%) are mostly zero (>{zero_threshold*100:.0f}% zeros).")
    return zero_indices

# Example with your test set
zero_patches = validate_dataset(test_dataset, zero_threshold=0.9, show_examples=True)
```

# Kalman Filter added to Astro U Net

```python
import os
import random
from datetime import datetime

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim
import wandb

# === Define your Kalman‚Äëaugmented U‚ÄëNet ===

class TorchKalmanDenoiser(nn.Module):
    def __init__(self, Q=1e-5, R=1e-2):
        super().__init__()
        self.Q = Q
        self.R = R

    def forward(self, meas: torch.Tensor, initial: torch.Tensor = None):
        meas = meas.float()
        X = initial if initial is not None else meas.clone()
        P = torch.ones_like(meas)

        P = P + self.Q
        K = P / (P + self.R)
        X = X + K * (meas - X)
        # P = (1 - K) * P  # we don't need P afterwards
        return X

class AstroKalmanUNet(nn.Module):
    def __init__(self, astrounet_model: AstroUNet=None,exp_injection=True, Q=1e-5, R=1e-2):
        super().__init__()
        # import your existing AstroUNet here
        if astrounet_model:
            self.unet   = astrounet_model
        else:
            self.unet   = AstroUNet(exp_injection=exp_injection)
        self.kalman = TorchKalmanDenoiser(Q=Q, R=R)

    def forward(self, x, exp_time=None):
        net_out   = self.unet(x, exp_time)
        filt_out  = self.kalman(net_out)
        return filt_out
```

```python

# --- Hyperparameters ---
BATCH_SIZE    = 4
NUM_EPOCHS    = 50
LEARNING_RATE = 1e-4
EXP_VAL       = 2.0
EPS           = 1e-8 

start_time = datetime.now().strftime("%Y%m%d_%H%M%S")
SAVE_DIR   = f"./checkpoints_astrounet/run_{start_time}/"
os.makedirs(SAVE_DIR, exist_ok=True)

# === W&B init ===
wandb.init(
    project="astrounet-jwst",
    name=f"kalman_run_{start_time}",
    config={
        "batch_size":    BATCH_SIZE,
        "epochs":        NUM_EPOCHS,
        "lr":            LEARNING_RATE,
        "exp_injection": True,
        "exp_val":       EXP_VAL,
        "kalman_Q":      1e-5,
        "kalman_R":      1e-2
    }
)

# === Data ===
train_ds = JWSTAstroDataset(txt_file="./astronet_training/train.txt")
val_ds   = JWSTAstroDataset(txt_file="./astronet_training/val.txt")

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=1,           shuffle=False, pin_memory=True)

# === Model, Loss, Optimizer ===
device    = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model     = AstroKalmanUNet(exp_injection=True, Q=1e-5, R=1e-2).to(device)
criterion = nn.L1Loss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# === Utility for KL divergence ===
def kl_divergence(a, b, bins=100):
    """Discrete KL(P||Q) for histograms of a and b."""
    # joint bins on combined range
    mn = min(a.min(), b.min())
    mx = max(a.max(), b.max())
    P, _ = np.histogram(a, bins=bins, range=(mn, mx), density=True)
    Q, _ = np.histogram(b, bins=bins, range=(mn, mx), density=True)
    P += EPS; Q += EPS
    return np.sum(P * np.log(P / Q))


best_val_psnr = 0.0

# === Training Loop ===
for epoch in range(1, NUM_EPOCHS + 1):
    model.train()
    running_loss = 0.0

    for noisy, clean in train_loader:
        noisy, clean = noisy.to(device), clean.to(device)
        exp_tensor   = torch.tensor(EXP_VAL, device=device)

        output = model(noisy, exp_tensor)
        loss   = criterion(output, clean)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    avg_train_loss = running_loss / len(train_loader)

    # === Validation ===
    model.eval()
    metrics = {"psnr": [], "ssim": [], "snr_noisy": [], "snr_out": [],
               "kl": [], "rel_flux_err": []}

    with torch.no_grad():
        for noisy, clean in val_loader:
            noisy, clean = noisy.to(device), clean.to(device)
            output = model(noisy, exp_tensor)

            # to numpy
            n_np = noisy.cpu().squeeze().numpy()
            o_np = output.cpu().squeeze().numpy()
            c_np = clean.cpu().squeeze().numpy()

            # PSNR & SSIM
            metrics["psnr"].append(psnr(c_np, o_np, data_range=1.0))
            metrics["ssim"].append(ssim(c_np, o_np, data_range=1.0))

            # SNR: mean(signal)/std(noise)
            sig_mean = c_np.mean()
            noise_noisy = (c_np - n_np).std()
            noise_out   = (c_np - o_np).std()
            metrics["snr_noisy"].append(sig_mean / (noise_noisy + EPS))
            metrics["snr_out"].append(sig_mean / (noise_out   + EPS))

            # KL divergence
            metrics["kl"].append(kl_divergence(c_np, o_np))

            # Relative flux error
            flux_c = c_np.sum()
            flux_o = o_np.sum()
            metrics["rel_flux_err"].append(abs(flux_o - flux_c) / (flux_c + EPS))

    # average metrics
    avg = {k: np.mean(v) for k, v in metrics.items()}

    # === Log to W&B ===
    wandb.log({
        "epoch":           epoch,
        "train/loss":      avg_train_loss,
        "val/psnr":        avg["psnr"],
        "val/ssim":        avg["ssim"],
        "val/snr_noisy":   avg["snr_noisy"],
        "val/snr_out":     avg["snr_out"],
        "val/kl_div":      avg["kl"],
        "val/rel_flux_err":avg["rel_flux_err"],
        "learning_rate":   optimizer.param_groups[0]["lr"]
    })

    # === Example images every 10 epochs ===
    if epoch % 10 == 0 or epoch == 1:
        noisy, clean = next(iter(val_loader))
        noisy, clean = noisy.to(device), clean.to(device)
        output = model(noisy, exp_tensor)
        
        noisy_np  = noisy.detach().cpu().squeeze().numpy()
        output_np = output.detach().cpu().squeeze().numpy()
        clean_np  = clean.detach().cpu().squeeze().numpy()


        grid = np.stack([noisy_np, output_np, clean_np], axis=-1)

        wandb.log({"examples":
            wandb.Image(grid, caption=f"Epoch {epoch}: Noisy/Denoised/Clean")
        })

    # === Checkpointing ===
    if avg["psnr"] > best_val_psnr:
        best_val_psnr = avg["psnr"]
        ckpt_path = os.path.join(SAVE_DIR, f"best_kalman_unet_{start_time}.pth")
        torch.save(model.state_dict(), ckpt_path)
        wandb.save(ckpt_path)
        print(f"‚úÖ Epoch {epoch}: Best PSNR={best_val_psnr:.3f}; saved {ckpt_path}")

    print(f"[{epoch}/{NUM_EPOCHS}] Train={avg_train_loss:.4f}  "
          f"PSNR={avg['psnr']:.2f}  SSIM={avg['ssim']:.3f}  "
          f"SNR‚Üë {avg['snr_noisy']:.2f}->{avg['snr_out']:.2f}  "
          f"KL={avg['kl']:.3f}  dF={avg['rel_flux_err']:.3f}")

# Finish
wandb.finish()
```
